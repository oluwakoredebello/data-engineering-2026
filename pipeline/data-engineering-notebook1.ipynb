```python
dtype = {
    "VendorID": "Int64",
    "passenger_count": "Int64",
    "trip_distance": "float64",
    "RatecodeID": "Int64",
    "store_and_fwd_flag": "string",
    "PULocationID": "Int64",
    "DOLocationID": "Int64",
    "payment_type": "Int64",
    "fare_amount": "float64",
    "extra": "float64",
    "mta_tax": "float64",
    "tip_amount": "float64",
    "tolls_amount": "float64",
    "improvement_surcharge": "float64",
    "total_amount": "float64",
    "congestion_surcharge": "float64"
}
```

```python
parse_dates = [
    "tpep_pickup_datetime",
    "tpep_dropoff_datetime"
]
```

```python
df_updated = pd.read_csv(
    url,
    dtype = dtype,
    parse_dates = parse_dates
)
```

```python
df_updated.head(1000)
```

```python
df_updated.dtypes #confirming updated datatypes
```

```python
from sqlalchemy import create_engine
engine = create_engine('postgresql://root:root@localhost:5432/ny_taxi')
```

```python
print(pd.io.sql.get_schema(df_updated, name='yellow_taxi_data', con=engine)) #making sure to input the right dataframe
```

```python
df_updated.head(0).to_sql(name = 'yellow_taxi_data', con = engine, if_exists = 'replace')
```

```python
#inserting data into PostgreSQL in chunks
df_iter = pd.read_csv(
    url,
    dtype = dtype, 
    parse_dates = parse_dates,
    iterator = True,
    chunksize = 100000,
)
```

```python
for df_chunk in df_iter:
    print(len(df_chunk))
```

```python
from tqdm.auto import tqdm
```

```python
for df_chunk in tqdm(df_iter):
    ...
    df_chunk.to_sql(name='yellow_taxi_data', con=engine, if_exists='append')
```
